```python
from anthropic import Anthropic
from google.colab import userdata

MODEL = 'claude-sonnet-4-0'
MAX_TOKENS = 100
API_KEY = userdata.get('API_KEY')
client = Anthropic(api_key=API_KEY)
```

```python
def user_message(messages, text):
  messages.append({'role': 'user', 'content': text})
  return messages

def assistant_message(messages, text):
  messages.append({'role': 'assistant', 'content': text})
  return messages

def chat(messages, prompt, system=None):
  user_message(messages, prompt)
  params = {
      'model':MODEL,
      'max_tokens':MAX_TOKENS,
      'messages':messages
  }
  if system:
    params['system'] = system

  response = client.messages.create(**params)

  answer = response.content[0].text
  assistant_message(messages, answer)
  return answer


messages = []
system = """"
You are a Senior Software Engineer with years of experience with python. You specialize
in creating good, concise code, that is short, and easy to read while being effective
solving the problem.
"""

while(True):
  user_prompt = input('> ')
  answer = chat(messages, user_prompt)
  print('-----\n')
  print(answer+'\n')
  print('-----\n')
```
